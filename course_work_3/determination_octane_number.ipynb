#### **Получившиеся данные**

```python
import pandas as pd

file_path = './structured_fuel_data.csv'
data = pd.read_csv(file_path)

print("Первые строки данных:")
display(data.head())
print("\nИнформация о данных:")
data.info()
```

```python
missing_data = data.isnull().sum()
print("Количество пропущенных значений в каждом столбце:")
print(missing_data[missing_data > 0])
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Тепловая карта пропущенных значений")
plt.show()
```

Судя по результатам анализа пропущенных значений, можно сделать следующие наблюдения:

Незначительное количество пропусков (4 пропущенных значения) присутствует почти во всех столбцах, за исключением нескольких, которые содержат значительно больше пропусков.
Существенные пропуски в столбцах, таких как 35/5_ароматика, 35/5_олефины, УПБКА1_ароматика, УПБКА1_олефины, и других, где отсутствует примерно 4421 запись. Эти столбцы имеют более чем 90% пропущенных данных.

В рамках предобработки данных были проведены следующие операции для обработки пропусков:
  * Замена пропущенных значений. Для столбцов, содержащих всего 4 пропущенных значения, была выполнена замена пропусков на медианные значения этих столбцов. Применение медианных значений позволяет минимизировать влияние выбросов и сохранить целостность данных, избегая искажений, которые могут возникнуть при использовании среднего значения.
  * Удаление столбцов с высоким процентом пропусков. Столбцы с более чем 90% пропусков были исключены из набора данных. Включение таких данных в анализ могло бы привести к неточным прогнозам и усложнить построение модели. Этот подход помогает сфокусироваться на более надежных и полных данных, что, в свою очередь, улучшает качество последующего анализа и прогностической модели.
  * Проверка на соответствие объема входящих и выходящих данных важна для оценки точности и эффективности производственного процесса.

```python
input_columns = [
    'УКК1_объем', 'ЛФ_объем', '35/5_объем', 'УПБКА1_объем', 'УПБКА2_объем',
    'БГБ_объем', 'ЛББГБ_объем', 'ЛЧ_объем', 'НК_объем'
]

# Вычисляем общий объем входящих потоков
data['total_input_volume'] = data[input_columns].sum(axis=1)

# Вычисляем разницу между входящим и выходящим объемом
# Предположим, что столбец 'вход' представляет собой итоговый объем
data['volume_discrepancy'] = data['total_input_volume'] - data['вход']

# Проверка несоответствий
print("Статистика несоответствий объема:")
print(data['volume_discrepancy'].describe())

# Вывод строк с наибольшими несоответствиями
print("\nПримеры строк с наибольшими несоответствиями:")
print(data[['total_input_volume', 'вход', 'volume_discrepancy']].sort_values(by='volume_discrepancy', ascending=False).head(10))
```

```python
# 1. Замена пропущенных значений медианой для числовых столбцов с 4 пропусками
columns_with_few_missing = data.columns[(data.isnull().sum() == 4) & (data.dtypes != 'object')]
data[columns_with_few_missing] = data[columns_with_few_missing].apply(lambda x: x.fillna(x.median()))

# 2. Удаление столбцов с более чем 90% пропущенных значений
threshold = 0.9  # Порог для удаления столбцов с пропусками (90%)
columns_to_drop = data.columns[data.isnull().mean() > threshold]
data_cleaned = data.drop(columns=columns_to_drop)
data_cleaned = data_cleaned.dropna(subset=['Timestamp'])

discrepancy_threshold = 0.05
data_cleaned = data_cleaned[abs(data_cleaned['volume_discrepancy']) <= discrepancy_threshold]

print(f"Оставшиеся данные после удаления строк с высоким несоответствием: {data_cleaned.shape[0]} строк")
print(data_cleaned[['total_input_volume', 'вход', 'volume_discrepancy']].head())

input_columns = [
    'УКК1_объем', 'ЛФ_объем', '35/5_объем', 'УПБКА1_объем', 'УПБКА2_объем',
    'БГБ_объем', 'ЛББГБ_объем', 'ЛЧ_объем', 'НК_объем'
]

# Вычисляем общий объем входящих потоков
data['total_input_volume'] = data[input_columns].sum(axis=1)

# Вычисляем разницу между входящим и выходящим объемом
# Предположим, что столбец 'вход' представляет собой итоговый объем
data['volume_discrepancy'] = data['total_input_volume'] - data['вход']

# Проверка несоответствий
print("Статистика несоответствий объема:")
print(data['volume_discrepancy'].describe())

# Вывод строк с наибольшими несоответствиями
print("\nПримеры строк с наибольшими несоответствиями:")
print(data[['total_input_volume', 'вход', 'volume_discrepancy']].sort_values(by='volume_discrepancy', ascending=False).head(10))

# Проверка результата
print("Столбцы, которые были удалены из-за высокого количества пропусков:")
print(columns_to_drop.tolist())

# Проверка, остались ли пропуски после обработки
remaining_missing = data_cleaned.isnull().sum().sum()
print(f"\nКоличество оставшихся пропущенных значений после обработки: {remaining_missing}")
```

#### **Анализ выбросов**
Цель обработки выбросов — улучшение качества данных и стабильности модели. Выбросы представляют собой наблюдения, значения которых значительно отличаются от большинства данных. Они могут возникать по разным причинам, включая ошибки измерения, а также редкие или экзогенные явления, которые не всегда относятся к общей тенденции данных. В процессе построения прогностической модели наличие выбросов может привести к смещению оценок, нарушению предположений о нормальности и снижению точности модели.

Методы обработки выбросов:

1. Выявление выбросов с использованием межквартильного размаха (IQR):
  * Межквартильный размах (Interquartile Range, IQR) — это разница между третьим (Q3) и первым квартилями (Q1) данных, которая характеризует центральное распределение данных.
  * Для наблюдений, которые выходят за пределы диапазона [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR], они считаются потенциальными выбросами.

2. Обрезание выбросов:
  * Значения, выходящие за указанный диапазон, будут обрезаны до максимального значения, соответствующего верхнему пределу (Q3 + 1.5 * IQR), или минимального значения, соответствующего нижнему пределу (Q1 - 1.5 * IQR).
  * Этот метод позволяет уменьшить влияние выбросов, сохраняя данные, близкие к основному диапазону, что помогает снизить вероятность смещения модели.

```python
import numpy as np
def trim_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Обрезаем значения, выходящие за пределы
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
data_cleaned = trim_outliers_iqr(data_cleaned, numeric_columns)
print("Обработка выбросов завершена. Пример обработанных данных:")
print(data_cleaned[numeric_columns].describe())
```

2. Нормализация — это метод масштабирования, при котором данные приводятся к диапазону [0, 1] или [-1, 1]. Нормализация полезна для данных с ненормальным распределением и может быть предпочтительна для моделей, таких как нейронные сети, где важен ограниченный диапазон значений.

Цель данного этапа — оценить эффективность базовых моделей машинного обучения для прогнозирования целевой переменной. Для этого мы выберем несколько моделей, обучим их на текущем наборе данных и проведем их оценку с использованием кросс-валидации. Такой подход позволяет объективно сравнить модели и определить, какая из них лучше соответствует характеристикам наших данных и демонстрирует высокую точность предсказаний.

Шаги для выполнения:

1. Выбор моделей:
  * Линейная регрессия: Обеспечивает анализ линейных зависимостей между признаками и целевой переменной, что позволяет оценить, подходит ли линейная модель для данных.
  * Случайный лес: Модель на основе ансамбля деревьев решений, способная выявлять более сложные, нелинейные зависимости, что делает её хорошим выбором для анализа данных со сложной структурой.

2. Метод оценки: кросс-валидация:
  * Для объективной оценки качества моделей используется метод кросс-валидации, который разбивает данные на несколько частей (фолдов) и позволяет оценить модель на различных подмножествах данных.

3. Метрики оценки:
  * Среднеквадратическая ошибка (MSE): Показывает среднее отклонение предсказанных значений от реальных, чувствительна к выбросам.
  * Коэффициент детерминации (R²): Определяет, какая часть дисперсии целевой переменной объясняется моделью, и указывает на качество подгонки модели к данным.

4. Подбор гиперпараметров: Если результаты одной из моделей будут удовлетворительными, можно будет провести настройку её гиперпараметров, чтобы повысить её точность. Настройка гиперпараметров, таких как количество деревьев и глубина для случайного леса, может улучшить производительность модели и повысить её обобщающую способность.

```python
input_streams = [
    'УКК1_объем', 'УКК1_ОЧИ', 'УКК1_ОЧМ', 'УКК1_ароматика', 'УКК1_олефины',
    'ЛФ_объем', 'ЛФ_ОЧИ', 'ЛФ_ОЧМ', 'ЛФ_ароматика', 'ЛФ_олефины',
    '35/5_объем', '35/5_ОЧИ', '35/5_ОЧМ',
    'УПБКА1_объем', 'УПБКА1_ОЧИ', 'УПБКА1_ОЧМ',
    'УПБКА2_объем', 'УПБКА2_ОЧИ', 'УПБКА2_ОЧМ',
    'БГБ_объем', 'БГБ_ОЧИ', 'БГБ_ОЧМ', 'БГБ_аромат', 'БГБ_олефы',
    'ЛББГБ_объем', 'ЛББГБ_ОЧИ', 'ЛББГБ_ОЧМ',
    'ЛЧ_объем', 'ЛЧ_ОЧИ', 'ЛЧ_ОЧМ',
    'НК_объем', 'НК_ОЧИ', 'НК_ОЧМ'
]

# Указываем целевую переменную — октановое число (например, ОЧИ)
target_column = '95_ОЧИ'  # или 'ОЧМ'

# Формируем набор данных с нужными столбцами
X = data_cleaned[input_streams]
y = data_cleaned[target_column]

# Проверка получившихся данных
print("Входные данные (9 потоков):")
print(X.head())
print("\nЦелевая переменная (октановое число):")
print(y.head())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42)
}

# Оценка моделей с помощью кросс-валидации
for model_name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)  # Извлекаем корень из MSE для получения RMSE
    print(f"{model_name} RMSE: {rmse_scores.mean():.2f} ± {rmse_scores.std():.2f}")
    r2_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    print(f"{model_name} R²: {r2_scores.mean():.2f} ± {r2_scores.std():.2f}\n")
```
